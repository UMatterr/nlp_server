```python
# 데이터 쌍 생성 함수
def create_data_pairs(col_list):
  pairs = []

  for i in range(len(col_list)):
    for j in range(i+1, len(col_list)):
        # Append each pair as a tuple to the list
        pairs.append((col_list[i], col_list[j]))
    return pairs

# 데이터 쌍 생성
col_list = ['formal', 'informal', 'enfp', 'joongding', 'gentle']
data_pairs = create_data_pairs(col_list)

# 데이터 쌍을 학습 데이터와 평가 데이터로 분리
train_data, test_data = train_test_split(data_pairs, test_size=0.1)
```
